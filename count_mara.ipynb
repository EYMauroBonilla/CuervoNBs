```python
import os
import pandas as pd

# Specify the directory containing the Excel files
directory = r'C:\Users\AW581PP\EY\Cuervo - General\3. EjecuciÃ³n\Perfilado de Datos\Materias Primas\Para Diccionario'

# Define the function to count valid entries, excluding blanks
def count_valid_values(column):
    # Filter the column to exclude '0', '0.00', '0.000' and blanks (empty strings)
    return column[(column != '0') & (column != '0.00') & (column != '0.000') & (column != '')].count()

# Loop through each file in the directory
for filename in os.listdir(directory):
    if filename.lower().endswith('.xlsx'):  # Check if the file is an Excel file
        # Construct the full file path
        file_path = os.path.join(directory, filename)
        
        # Read the Excel file
        df = pd.read_excel(file_path, dtype=str)
        
        # Drop the first row from the dataframe
        df = df.iloc[1:]
        
        # Apply the counting function across all columns starting from the third row, which is now row 1
        valid_counts = df.iloc[1:].apply(count_valid_values)
        
        # Convert counts to integers explicitly
        valid_counts = valid_counts.astype(int)
        
        # Use counts as the new header and remove the old header which is now incorrect
        new_header = valid_counts.index  # Store new header names which are the original column names
        df = df.iloc[1:]  # Drop old header now residing at row 0 of df after initial drop
        df.columns = new_header  # Set new column names to the existing column names
        
        # Insert the counts as a new DataFrame row at the top
        counts_row = pd.DataFrame([valid_counts], columns=new_header)
        df = pd.concat([counts_row, df], ignore_index=True)
        
        # Define the new file name
        new_file_path = os.path.join(directory, 'count_' + filename)
        
        # Write the results to a new Excel file
        df.to_excel(new_file_path, index=False)
        
        # Print completion message
        print(f"Done with {filename}")

print("All files have been processed and updated copies created.")
```
